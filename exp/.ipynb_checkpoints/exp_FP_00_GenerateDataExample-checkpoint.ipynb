```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Generate data
id_values = np.arange(1, 101)  # Equivalent to 1:100 in R
y_values = np.random.normal(size=100)  # Equivalent to rnorm(100) in R
```

```python
# Save data to a CSV file
data = pd.DataFrame({'id': id_values, 'y': y_values})
data.to_csv('../dat/data.csv', index=False)
```

```python
# Plot
plt.scatter(id_values, y_values)  # Equivalent to plot(y ~ id, dat) in R
plt.xlabel('id')
plt.ylabel('y')
plt.title('Scatter plot of y vs. id')

# Save the figure as a PDF
plt.savefig('../doc/fig/scatter_plot.pdf')
```
